```python
import sys
from collections import Counter

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


class Solution:
    def mins(self, a, d):
        n = len(a)
        dp = [float("inf")] * (n + 1)
        dp[n] = 0
        for i in range(n - 1, -1, -1):
            for j in range(i + 1, n + 1):
                if a[j - 1] > a[i - 1]:
                    dp[i] = min(dp[i], (d[a[i - 1]] - 1) * a[j - 1] + a[i - 1] + dp[j])
        return dp[1]

    def solve(self):
        t = int(input())
        for _ in range(t):
            n = int(input())
            a = list(map(int, input().split()))
            c = Counter(a)
            if c[0] <= 1:
                print(0)
                continue
            mex = 0
            while mex in c:
                mex += 1
            e = sorted(list(set(a)))
            e.append(mex)
            c[mex] = 1
            print(self.mins(e, c))


s = Solution()
s.solve()
```

```python
import sys
from collections import Counter

input = sys.stdin.readline

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")


class Solution:
    def mins(self, a, d):
        def solve(begin, prev):
            mins = float("inf")
            if begin == len(a):
                return 0
            for i in range(begin, len(a)):
                if a[i] > prev:
                    mins = min(mins, (d[prev] - 1) * a[i] + prev + solve(i + 1, a[i]))
            return mins

        return solve(0, 0)


s = Solution()
t = int(input())
for _ in range(t):
    n = int(input())
    a = list(map(int, input().split()))
    c = Counter(a)
    if c[0] <= 1:
        print(0)
        continue
    mex = 0
    while mex in c:
        mex += 1
    e = sorted(list(set(a)))
    e.append(mex)
    c[mex] = 1
    print(s.mins(e, c))
```
