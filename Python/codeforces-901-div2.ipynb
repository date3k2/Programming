```python
import sys
from collections import Counter

input = sys.stdin.readline

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")


class Solution:
    def mins(self, a, d):
        def solve(begin, prev):
            mins = float("inf")
            if begin == len(a):
                return 0
            for i in range(begin, len(a)):
                if a[i] > prev:
                    mins = min(mins, (d[prev] - 1) * a[i] + prev + solve(i + 1, a[i]))
            return mins

        return solve(0, 0)


s = Solution()
t = int(input())
for _ in range(t):
    n = int(input())
    a = list(map(int, input().split()))
    c = Counter(a)
    if c[0] <= 1:
        print(0)
        continue
    mex = 0
    while mex in c:
        mex += 1
    e = sorted(list(set(a)))
    e.append(mex)
    c[mex] = 1
    print(s.mins(e, c))
```

```python
from functools import lru_cache


@lru_cache()
def fibo(n):
    if n <= 1:
        return n
    return fibo(n - 1) + fibo(n - 2)


print(fibo(int(input()) - 1))
```

```python
def gen_bin_pem(n):
    for i in range(0, 2):
        if n == 1:
            yield str(i)
        else:
            for j in gen_bin_pem(n - 1):
                if i == 1 and j[0] == "1":
                    continue
                yield str(i) + j


for i in gen_bin_pem(int(input())):
    print(i)
```

```python
class Permutation:
    def __init__(self, n) -> None:
        self.a = []
        self.chosen = [False] * n
        self.n = n

    def permutation(self):
        if len(self.a) == self.n:
            print(*self.a)
        else:
            for i in range(self.n):
                if self.chosen[i]:
                    continue
                self.chosen[i] = True
                self.a.append(i + 1)
                self.permutation()
                self.chosen[i] = False
                self.a.pop()


Permutation(int(input())).permutation()
```

```python
MOD = 10**9 + 7


def comb(n, k):
    if k == 0 or k == n:
        return 1
    res = 1
    for i in range(k):
        res = res * (n - i) // (i + 1)
    return res % MOD


k, n = map(int, input().split())
print(comb(n, k))
```

```python
from typing import List
import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")

input = sys.stdin.readline


def countSudoku(matrix: List[List[int]]) -> int:
    def check(x, y, k):
        for i in range(9):
            if matrix[x][i] == k or matrix[i][y] == k:
                return False
        for i in range(3):
            for j in range(3):
                if matrix[x // 3 * 3 + i][y // 3 * 3 + j] == k:
                    return False
        return True

    def solve(x, y):
        if x == 9:
            return True
        if y == 9:
            return solve(x + 1, 0)
        if matrix[x][y] != 0:
            return solve(x, y + 1)
        for i in range(1, 10):
            if check(x, y, i):
                matrix[x][y] = i
                if solve(x, y + 1):
                    return True
                matrix[x][y] = 0
        return False

    solve(0, 0)
    # return matrix
matrix = [list(map(int, input().split())) for _ in range(9)]
print(countSudoku(matrix))
```

```python
class Solution:
    def isValidSudoku(self, board: List[List[str]]) -> bool:
        def check(x, y, k):
            for i in range(9):
                if board[x][i] == k or board[i][y] == k:
                    return False
            for i in range(3):
                for j in range(3):
                    if board[x // 3 * 3 + i][y // 3 * 3 + j] == k:
                        return False
            return True

        def solve(x, y):
            if x == 9:
                return True
            if y == 9:
                return solve(x + 1, 0)
            if board[x][y] != ".":
                return solve(x, y + 1)
            for i in range(1, 10):
                if check(x, y, str(i)):
                    board[x][y] = str(i)
                    if solve(x, y + 1):
                        return True
                    board[x][y] = "."
            return False

        solve(0, 0)
```
