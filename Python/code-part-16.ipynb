```python
import sys

sys.stdin = open(
    r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(
    r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


class Solution:
    def compressString(self, S: str) -> str:
        if len(S) == 0:
            return ""
        res = ""
        count = 1
        for i in range(len(S) - 1):
            if S[i] == S[i + 1]:
                count += 1
            else:
                res += S[i] + str(count) if count > 1 else S[i]
                count = 1
        res += S[-1] + str(count) if count > 1 else S[-1]
        return res if len(res) < len(S) else S


t = int(input())
s = Solution()
for _ in range(t):
    S = input().rstrip()
    print(s.compressString(S))
```

```python
import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


class Solution:
    def isCenterOfStarGraph(self, edges, node):
        for l in edges:
            if l[0] != node or l[1] == node:
                return "false"
        return "true"


t = int(input())
s = Solution()
for _ in range(t):
    edges = eval(input())
    node = int(input())
    print(s.isCenterOfStarGraph(edges, node))
```

```python
import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


class BinaryTreeNode:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None


class Solution:
    def insertNode(self, root, data):
        if root is None:
            root = BinaryTreeNode(data)
        elif data <= root.data:
            root.left = self.insertNode(root.left, data)
        else:
            root.right = self.insertNode(root.right, data)
        return root

    def printAllLeaves(self, root):
        if root is None:
            return
        if root.left is None and root.right is None:
            print(root.data, end=" ")
        self.printAllLeaves(root.left)
        self.printAllLeaves(root.right)


t = int(input())
for _ in range(t):
    n = int(input())
    a = list(map(int, input().split()))
    s = Solution()
    root = None
    for i in range(n):
        root = s.insertNode(root, a[i])
    s.printAllLeaves(root)
    print()
```

```python
import sys
from collections import defaultdict

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline
t = int(input())
for _ in range(t):
    n = int(input())
    a = list(map(int, input().split()))
    if n == 1:
        print(0)
        continue
    d = defaultdict(int)
    for i in a:
        d[i] += 1
    ans = 0
    for i in range(n - 2, -1, -1):
        if d[a[i]] == 0:
            a[i] = 0
            continue
        if a[i] > a[i + 1]:
            d[a[i]] = 0
            a[i] = 0
            ans += 1
    if d[a[-1]] == 0 and a[-2] > 0:
        ans += 1
    print(ans)
```
