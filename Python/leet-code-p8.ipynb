```python
from typing import List
from collections import defaultdict


class Solution:
    def countNicePairs(self, nums: List[int]) -> int:
        pal = 0
        f = defaultdict(int)
        for num in nums:
            s = str(num)
            n = len(s)
            sub = []
            check = True
            for i in range(n // 2):
                diff = int(s[i]) - int(s[n - i - 1])
                if diff:
                    check = False
                sub.append(str(diff))
            if check:
                pal += 1
            else:
                sub = "".join(sub)
                f[(sub, n)] += 1

        ans = pal * (pal - 1) // 2
        for val in f.values():
            ans += val * (val - 1) // 2
        return ans % 1000000007


nums = [42, 11, 1, 97]
nums = [13, 10, 35, 24, 76]
nums = [131, 131, 131]
nums = [43, 71, 24, 47, 36, 99, 60, 83, 51, 44]
nums = [
    34741,
    86419,
    43784,
    70003,
    13896,
    67799,
    36762,
    80803,
    69006,
    35996,
    1963,
    78009,
    80610,
    80342,
    63169,
    87967,
    38893,
    94455,
    23987,
    41052,
    74154,
    26449,
    84207,
    99184,
    16976,
    61619,
    60605,
    3588,
    72461,
    39369,
    65776,
    12063,
    3871,
    71664,
    19128,
    69607,
    77799,
    79407,
    47319,
    81112,
    21204,
    45240,
    42308,
    55148,
    87450,
    10303,
    39546,
    858,
    76843,
    98801,
    94111,
    86522,
    15998,
    875,
    31658,
    41600,
    23629,
    58414,
    19577,
    18428,
    50586,
    89583,
    64408,
    10629,
    87864,
    2061,
    31242,
    1308,
    68261,
    66053,
    36126,
    91907,
    59263,
    7095,
    49624,
    99486,
    57475,
    49493,
    9594,
    74495,
    84333,
    44560,
    14546,
    76406,
    28489,
    41131,
    83307,
    5358,
    10422,
    68306,
    42132,
    60276,
    95348,
    55761,
    94163,
    61656,
    74908,
    83159,
    20521,
    79925,
    45557,
    88386,
    47417,
    88895,
    17656,
    17734,
    96471,
    38739,
    78568,
    44871,
    87306,
    84403,
    33175,
    97730,
    24102,
    3996,
    81201,
    45727,
    79906,
    46784,
    66053,
    95612,
    27857,
    61293,
    72828,
    62079,
    60499,
    16038,
    74414,
    68973,
    69924,
    24395,
    80385,
    88735,
    89754,
    73829,
    58268,
    23691,
    15630,
    17267,
    54205,
    30440,
    31052,
    58028,
    16953,
    66231,
    85608,
    48349,
    18570,
    16926,
    15193,
    38595,
    94793,
    56912,
    24485,
    45952,
    79126,
    28751,
    6413,
    89248,
    81157,
    43568,
    41712,
    5616,
    86849,
    25707,
    85757,
    19034,
    30397,
    61215,
    72340,
    78361,
    74778,
    45055,
    46299,
    9570,
    18970,
    97967,
    22239,
    25793,
    83826,
    27735,
    16433,
    85146,
    48928,
    8738,
    5766,
    26052,
    46605,
    59034,
    92625,
    84960,
    28810,
    73832,
    7592,
    17505,
    16619,
    70061,
    45252,
    92287,
]
nums = [120, 32]
Solution().countNicePairs(nums)
```
