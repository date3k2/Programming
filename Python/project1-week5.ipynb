```python
import sys

sys.stdin = open("input.txt", "r")
sys.stdout = open("output.txt", "w")
input = sys.stdin.readline


# --- Check if a graph is Hamiltonian or not ---
class HamiltonianGraph:
    def __init__(self, n, adj, v) -> None:
        self.n = n
        self.adj = adj
        self.visited = set()
        self.v = v

    def check(self, u):
        if len(self.visited) == self.n and u == self.v:
            return True
        for v in self.adj[u]:
            if v not in self.visited:
                self.visited.add(v)
                if self.check(v):
                    return True
                self.visited.remove(v)


if __name__ == "__main__":
    t = int(input())
    for _ in range(t):
        n, m = map(int, input().split())
        adj = [[] for _ in range(n)]
        for _ in range(m):
            a, b = map(int, input().split())
            adj[a - 1].append(b - 1)
            adj[b - 1].append(a - 1)
        for i in range(n):
            obj = HamiltonianGraph(n, adj, i)
            if obj.check(i):
                print(1)
                break
        else:
            print(0)
```

```python
from typing import List


class Solution:
    def numFactoredBinaryTrees(self, a: List[int]) -> int:
        a.sort()
        dp = {i: 1 for i in a}
        MOD = 1000000007
        for i in a:
            for j in a:
                if i % j == 0 and i // j in dp:
                    dp[i] += dp[j] * dp[i // j] if i // j != j else dp[j] * dp[i // j]
                    dp[i] %= MOD
        return sum(dp.values()) % MOD


arr = [45,42,2,18,23,1170,12,41,40,9,47,24,33,28,10,32,29,17,46,11,759,37,6,26,21,49,31,14,19,8,13,7,27,22,3,36,34,38,39,30,43,15,4,16,35,25,20,44,5,48]
print(Solution().numFactoredBinaryTrees(arr))
```

```python
class Solution:
    def longestPalindrome(self, s: str):
        def expand(i, j):
            while i >= 0 and j < len(s) and s[i] == s[j]:
                i -= 1
                j += 1
            return s[i + 1 : j]

        res = ""
        for i in range(len(s)):
            res = max(expand(i, i), expand(i, i + 1), res, key=len)
        return res


s = "ihccaduabejbhreqkxxdvkmbhnxofhgddhdlflyswrvdqeibqdsntbjtdqnydxakghsuxmxksjpgbuvbhjdlmcvtwsdbwfmqfslk"
# s = "cbbdnva"
print(Solution().longestPalindrome(s))
```

```python
import sys
from heapq import heappush, heappop
from collections import defaultdict

input = sys.stdin.readline
# --- Dijkstra's Algorithm ---


class Graph:
    def __init__(self, n: int):
        self.graph = defaultdict(list)
        self.n = n

    def addEdge(self, u, v, w):
        self.graph[u].append((v, w))

    def dijkstra(self, start, end):
        visited = [False] * self.n
        dist = [float("inf")] * self.n
        dist[start] = 0
        pq = [(0, start)]
        while pq:
            _, node = heappop(pq)
            if visited[node]:
                continue
            visited[node] = True
            if node == end:
                break
            for v, w in self.graph[node]:
                if dist[v] > dist[node] + w:
                    dist[v] = dist[node] + w
                    heappush(pq, (dist[v], v))
        return dist[end] if dist[end] != float("inf") else -1


if __name__ == "__main__":
    n, m = map(int, input().split())
    g = Graph(n)
    for _ in range(m):
        u, v, w = map(int, input().split())
        g.addEdge(u - 1, v - 1, w)
    s, t = map(int, input().split())
    print(g.dijkstra(s - 1, t - 1))
```

```python
import sys
from collections import defaultdict

# sys.stdin = open("input.txt", "r")
# sys.stdout = open("output.txt", "w")
input = sys.stdin.readline
# --- Shortest Path between all pairs of nodes ---


class Graph:
    def __init__(self, n: int):
        self.graph = defaultdict(list)
        self.n = n

    def addEdge(self, u, v, w):
        self.graph[u].append((v, w))

    def floydWarshall(self):
        dist = [[float("inf")] * self.n for _ in range(self.n)]
        for i in range(self.n):
            dist[i][i] = 0
        for u in range(self.n):
            for v, w in self.graph[u]:
                dist[u][v] = w
        for k in range(self.n):
            for i in range(self.n):
                for j in range(self.n):
                    dist[i][j] = min(dist[i][j], dist[i][k] + dist[k][j])
        return dist


if __name__ == "__main__":
    n, m = map(int, input().split())
    g = Graph(n)
    for _ in range(m):
        u, v, w = map(int, input().split())
        g.addEdge(u - 1, v - 1, w)
    dist = g.floydWarshall()
    for i in range(n):
        for j in range(n):
            print(dist[i][j] if dist[i][j] != float("inf") else -1, end=" ")
        print()
```

```python
import sys
from collections import defaultdict, deque

# sys.stdin = open("input.txt", "r")
# sys.stdout = open("output.txt", "w")
input = sys.stdin.readline
# --- Maximum Flow in a Graph ---


class Graph:
    def __init__(self, n: int):
        self.n = n
        self.parent = [-1] * n
        self.edge = defaultdict(int)

    def addEdge(self, u, v, capacity):
        self.edge[(u, v)] = capacity

    def bfs(self, source, sink):
        visited = [False] * self.n
        q = deque()
        q.append(source)
        visited[source] = True
        while q:
            u = q.popleft()
            for i in range(self.n):
                if not visited[i] and self.edge[(u, i)] > 0:
                    visited[i] = True
                    self.parent[i] = u
                    if i == sink:
                        return True
                    q.append(i)

        return False

    def fordFulkerson(self, source, sink):
        maxFlow = 0
        while self.bfs(source, sink):
            pathFlow = float("inf")
            v = sink
            while v != source:
                u = self.parent[v]
                pathFlow = min(pathFlow, self.edge[(u, v)])
                v = u
            maxFlow += pathFlow
            v = sink
            while v != source:
                u = self.parent[v]
                self.edge[(u, v)] -= pathFlow
                self.edge[(v, u)] += pathFlow
                v = u
        return maxFlow


if __name__ == "__main__":
    n, m = map(int, input().split())
    g = Graph(n)
    s, t = map(int, input().split())
    for _ in range(m):
        u, v, w = map(int, input().split())
        g.addEdge(u - 1, v - 1, w)
    print(g.fordFulkerson(s - 1, t - 1))
```
