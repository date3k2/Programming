```python
import sys

sys.stdin = open("input.txt", "r")
sys.stdout = open("output.txt", "w")
input = sys.stdin.readline
# ------------------------------
from bisect import bisect_left

t = int(input())
for _ in range(t):
    n = int(input())
    bits = input().rstrip()
    idx_0 = []
    idx_1 = []
    for i, s in enumerate(bits):
        idx_0.append(i) if s == "0" else idx_1.append(i)
    len1 = len(idx_1)
    ans = 0
    for i in idx_0[::-1]:
        ans += len1 - bisect_left(idx_1, i)
        print(ans, end=" ")
    for _ in range(len1):
        print(-1, end=" ")
    print()
```

```python
import sys

sys.stdin = open("input.txt", "r")
sys.stdout = open("output.txt", "w")
input = sys.stdin.readline
# ------------------------------

t = int(input())
for _ in range(t):
    code = input().rstrip()
    ans = 4
    old = 1
    for i in code:
        if (int(i) == 0 or old == 0) and int(i) + old != 0:
            ans += 10 - max(int(i), old)
        else:
            ans += abs(int(i) - old)
        old = int(i)
    print(ans)
```

```python
import sys

sys.stdin = open("input.txt", "r")
sys.stdout = open("output.txt", "w")
input = sys.stdin.readline
# ------------------------------
from collections import Counter

t = int(input())
for _ in range(t):
    n, k = map(int, input().split())
    s = input().rstrip()
    c = Counter(s)
    nums = 0
    for i in c:
        if c[i] % 2:
            if k <= 0:
                break
            k -= 1
            c[i] -= 1
    if k > 0:
        print("YES")
    else:
        count_odd = 0
        for i in c:
            count_odd += c[i] % 2
            if count_odd > 1:
                print("NO")
                break
        else:
            print("YES")
```

```python
import sys

sys.stdin = open("input.txt", "r")
sys.stdout = open("output.txt", "w")
input = sys.stdin.readline
# ------------------------------

t = int(input())
for _ in range(t):
    n, k = map(int, input().split())
    a = list(map(int, input().split()))
    if k == 2:
        for i in a:
            if i % 2 == 0:
                print(0)
                break
        else:
            print(1)
    elif k == 3:
        is_mod3_1 = is_mod3_2 = 5
        for i in a:
            if i % 3 == 0:
                print(0)
                break
            elif i % 3 == 1:
                is_mod3_1 = 2
            else:
                is_mod3_2 = 1
        else:
            print(min(is_mod3_1, is_mod3_2))
    elif k == 4:
        count_even = 0
        is_mod4_3 = 5
        for i in a:
            if i % 4 == 0:
                print(0)
                break
            elif i % 2 == 0:
                count_even += 1
                if count_even >= 2:
                    print(0)
                    break
            elif i % 4 == 3:
                is_mod4_3 = 1
        else:
            print(min(2 - count_even, is_mod4_3))
    else:
        is_mod5_1 = is_mod5_2 = is_mod5_3 = is_mod5_4 = 5
        for i in a:
            if i % 5 == 0:
                print(0)
                break
            elif i % 5 == 1:
                is_mod5_1 = 4
            elif i % 5 == 2:
                is_mod5_2 = 3
            elif i % 5 == 3:
                is_mod5_3 = 2
            else:
                is_mod5_4 = 1
        else:
            print(min(is_mod5_1, is_mod5_2, is_mod5_3, is_mod5_4))
```

```python
import sys

sys.stdin = open("input.txt", "r")
sys.stdout = open("output.txt", "w")
input = sys.stdin.readline
# ------------------------------
from math import log2, ceil

t = int(input())
for _ in range(t):
    n = int(input())
    a = list(map(int, input().split()))
    ans = count = old_count = 0
    for i in range(n - 1):
        log = ceil(log2(a[i] / a[i + 1]))
        if log + old_count < 0:
            old_count = 0
            continue
        old_count = old_count + log
        ans += old_count
    print(ans)
```
