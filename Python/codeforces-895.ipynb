```python
import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


def is_prime(n):
    if n < 2:
        return -1
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return i
    return -1


t = int(input())
for _ in range(t):
    l, r = map(int, input().split())
    if l < r:
        while r % 2 != 0:
            r -= 1
        if r == 2:
            print(-1)
        else:
            print(2, r - 2)
    else:
        if l == 2:
            print(-1)
        elif l % 2 == 0:
            print(l - 2, 2)
        elif is_prime(l) == -1:
            print(-1)
        else:
            print(is_prime(l), l - is_prime(l))
```

```python
import sys
from math import lcm

input = sys.stdin.readline


t = int(input())
for _ in range(t):
    n, x, y = map(int, input().split())
    range_x = n // x - n // lcm(x, y)
    range_y = n // y - n // lcm(x, y)
    sum1 = (n + n - range_x + 1) * range_x // 2
    sum2 = range_y * (range_y + 1) // 2
    print(sum1 - sum2)
```

```python
import sys

sys.stdin = open(
    r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(
    r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


t = int(input())
for _ in range(t):
    n = int(input())
    graph = list(map(int, input().split()))
    c = list(map(int, input().split()))
    deg_in = [0] * (n + 1)
    state = [0] * (n + 1)
    for i in range(n):
        deg_in[graph[i]] += 1
    q = [i for i in range(1, n + 1) if deg_in[i] == 0]
    graph = [0] + graph
    while q:
        u = q.pop()
        print(u, end=" ")
        state[u] = 1
        deg_in[graph[u]] -= 1
        if deg_in[graph[u]] == 0:
            q.append(graph[u])
    for i in range(1, n + 1):
        if state[i] == 0:
            u = i
            mins = c[i - 1]
            v = 0
            cycle = []
            while state[u] == 0:
                cycle.append(u)
                if c[u - 1] < mins:
                    mins = c[u - 1]
                    v = len(cycle) - 1
                state[u] = 1
                u = graph[u]
            print(*cycle[v + 1:], *cycle[: v + 1], end=" ")

    print()
```

```python
import sys

sys.stdin = open(
    r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(
    r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


t = int(input())
for _ in range(t):
    n, k, a, b = map(int, input().split())
    xy = [tuple(map(int, input().split())) for _ in range(n)]
    a -= 1
    b -= 1
    dis_a = 0 if 0 <= a <= k - 1 else float("inf")
    dis_b = 0 if 0 <= b <= k - 1 else float("inf")
    for i in range(k):
        dis_a = min(dis_a, abs(xy[a][0] - xy[i][0]) + abs(xy[a][1] - xy[i][1]))
        dis_b = min(dis_b, abs(xy[b][0] - xy[i][0]) + abs(xy[b][1] - xy[i][1]))
    print(min(dis_a + dis_b, abs(xy[a][0] -
          xy[b][0]) + abs(xy[a][1] - xy[b][1])))
```

```python
import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


t = int(input())
for _ in range(t):
    n, m = map(int, input().split())
    base = list(range(m))
    if m == 1:
        print(0)
        for _ in range(n):
            print(0)
        continue
    t = min(n, m)
    matrix = [base[i:] + base[:i] for i in range(t)]
    for i in range(len(matrix), n):
        matrix.append(matrix[i % t])
    print(min(n + 1, m))
    for i in matrix:
        print(*i)
```
