```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(.1, 50, 1000)
y = np.log(x) + np.cos(x)
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title(r'$ \ln{x} + \cos{x} $')
x = np.linspace(0, np.pi/2, 100)
plt.show()
```
