```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(.1, 50, 1000)
y = np.log(x) + np.cos(x)
fig1 = plt.figure()
axes1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x, y, 'r')
axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title(r'$ \ln{x} + \cos{x} $')
axes1.legend(['ln(x) + cos(x)'])
x = np.linspace(0, np.pi/2, 100)
plt.show()
```

```python
x = np.arange(11)
y = x * 2
z= x**2
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.9,0.9])
axes2 = fig.add_axes([0.2,0.8,0.1,0.1])

axes1.plot(x, y, 'g')
axes1.set_xlabel('X Label') # Notice the use of set_ to begin methods
axes1.set_ylabel('y Label')
axes1.text(0.5, 0.5, r'$y = 2x$', fontsize=14, color='blue')
axes1.set_title('Plot Title') # put ";" if you don't want to see the line of text below!\

axes2.plot(x,z,'b')
axes2.set_xlabel('x_inset')
axes2.set_ylabel('y_inset')
axes2.set_title('inset')
axes2.set_xlim(1,2)
axes2.set_ylim(0,5)

plt.show()
```

```python
x = np.linspace(0, 5, 20)
plt.fill_between(x, x**2, x**3, color='blue', alpha=0.5)
plt.show()
```

```python
import seaborn as sns
sns.set_style('whitegrid')
tips = sns.load_dataset('tips')
print(tips.head())
sns.histplot(tips['total_bill'], kde=True)
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='reg')
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde',fill = True)
plt.show()
```

```python
sns.pairplot(tips,hue='sex')
plt.show()
```

```python
sns.stripplot(x='day',y='total_bill',data=tips,hue = 'sex').legend(loc=1)
plt.show()
sns.swarmplot(x='day',y='total_bill',data=tips,hue = 'sex')
```

```python
sns.boxplot(data = tips,x = 'day',y='total_bill',hue='sex',palette='viridis')
```

```python
sns.pointplot(data=tips,x = 'day',y='total_bill',hue='smoker')
sns.set_theme(style="ticks",color_codes=True)
plt.show()
sns.countplot(data=tips,x='day',hue='smoker')
```
