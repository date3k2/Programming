## Linear Regression


```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.linear_model import LinearRegression, HuberRegressor
```

```python
X = np.array([[147, 150, 153, 158, 163, 165, 168, 170, 173, 175, 178, 180,
               183, 250]]).T
Y = np.array([49, 50, 51, 54, 58, 59, 60, 62, 63, 64, 66, 67, 68, 80])
plt.title('Regression', fontdict={'font': 'Fira Code', 'size': '25'})
model1 = LinearRegression()
model1.fit(X, Y)
plt.scatter(X, Y, marker="o")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.plot(X, model1.predict(X), color='red',
         linewidth=2, label='Linear Regression')

huber = HuberRegressor(max_iter=301, epsilon=1.2, alpha=0.3).fit(X, Y)
plt.grid()
plt.plot(X, huber.predict(X), color='green', label='Huber Regression')
plt.legend()
plt.show()
print('Coefficients:', model1.coef_[0], huber.coef_)
print('Intercept:', model1.intercept_, huber.intercept_)

# one = np.ones(X.shape)
# # x = np.concatenate((X,one),axis=1)
# # one = np.ones((X.shape[0], 1))
# Xbar = np.concatenate((one, X), axis = 1)
# print(Xbar)
# A = np.dot(Xbar.T, Xbar)
# print(A)
```
