```python
import sys
from collections import deque

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline
a, b, c = map(int, input().split())
filled = {(0, 0): 0}
states = deque([(0, 0)])
while states:
    x, y = states.popleft()
    if x == c or y == c:
        print(filled[(x, y)])
        break
    adj = set(
        [
            (a, y),
            (x, b),
            (0, y),
            (x, 0),
            (x + min(y, a - x), y - min(y, a - x)),
            (x - min(x, b - y), y + min(x, b - y)),
        ]
    )
    for i in adj:
        if i not in filled:
            states.append(i)
            filled[i] = filled[(x, y)] + 1
else:
    print(-1)
```

```python
class TreeNode:
    def __init__(self, val):
        self.val = val
        self.leftMostChild = None
        self.rightSibling = None


class Tree:
    def __init__(self, val):
        self.root = TreeNode(val)
        self.val_to_node = {val: self.root}
        self.parent = {val: None}

    def insert(self, u, v):
        if u in self.val_to_node or v not in self.val_to_node:
            return
        node = TreeNode(u)
        self.val_to_node[u] = node
        if not self.val_to_node[v].leftMostChild:
            self.val_to_node[v].leftMostChild = node
        else:
            temp = self.val_to_node[v].leftMostChild
            while temp.rightSibling:
                temp = temp.rightSibling
            temp.rightSibling = node
        self.parent[u] = v

    def preOrder(self):
        def dfs(node):
            if not node:
                return
            print(node.val, end=" ")
            dfs(node.leftMostChild)
            dfs(node.rightSibling)

        dfs(self.root)

    def postOrder(self):
        def dfs(node):
            if not node:
                return
            dfs(node.leftMostChild)
            print(node.val, end=" ")
            dfs(node.rightSibling)

        dfs(self.root)

    def inOrder(self):
        check = {}

        def dfs(node):
            if not node:
                return
            dfs(node.leftMostChild)
            if self.parent[node.val] and self.parent[node.val] not in check:
                if node.val not in check:
                    print(node.val, self.parent[node.val], end=" ")
                    check[node.val] = 1
                else:
                    print(self.parent[node.val], end=" ")
                check[self.parent[node.val]] = 1
            elif node.val not in check:
                print(node.val, end=" ")
            dfs(node.rightSibling)

        dfs(self.root)


import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline
while True:
    s = input().rstrip()
    if s == "*":
        break
    s = s.split()
    if s[0] == "MakeRoot":
        tree = Tree(s[1])
    elif s[0] == "Insert":
        tree.insert(s[1], s[2])
    elif s[0] == "PreOrder":
        tree.preOrder()
        print()
    elif s[0] == "InOrder":
        tree.inOrder()
        print()
    elif s[0] == "PostOrder":
        tree.postOrder()
        print()
```

```python
class TreeNode:
    def __init__(self, name):
        self.name = name
        self.children = []


class FamilyTree:
    def __init__(self):
        self.names = {}

    def insert(self, child: str, parent: str):
        if child not in self.names:
            self.names[child] = TreeNode(child)
        if parent not in self.names:
            self.names[parent] = TreeNode(parent)
        self.names[parent].children.append(self.names[child])

    def descendants(self, name: str) -> int:
        if name not in self.names:
            return 0
        count = -1
        stack = [self.names[name]]
        while stack:
            node = stack.pop()
            count += 1
            for child in node.children:
                stack.append(child)
        return count

    def generation(self, name: str) -> int:
        if name not in self.names:
            return 0
        count = -1
        stack = [self.names[name]]
        while stack:
            count += 1
            temp = []
            for node in stack:
                temp += node.children
            stack = temp
        return count


import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline
ft = FamilyTree()
while True:
    s = input().rstrip()
    if s == "***":
        break
    child, parent = s.split()
    ft.insert(child, parent)
while True:
    s = input().rstrip()
    if s == "***":
        break
    cmd, name = s.split()
    if cmd == "descendants":
        print(ft.descendants(name))
    else:
        print(ft.generation(name))
```

```python
class TreeNode:
    def __init__(self, val: int, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


class BinarySearchTree:
    def __init__(self, val: int):
        self.root = TreeNode(val)

    def insert(self, node: TreeNode, val: int):
        if node.val == val:
            return
        if val < node.val:
            if not node.left:
                node.left = TreeNode(val)
            else:
                self.insert(node.left, val)
        else:
            if not node.right:
                node.right = TreeNode(val)
            else:
                self.insert(node.right, val)

    def preOrder(self):
        def dfs(node: TreeNode):
            if not node:
                return
            print(node.val, end=" ")
            dfs(node.left)
            dfs(node.right)

        dfs(self.root)


import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline
_, key = input().split()
bst = BinarySearchTree(int(key))
while True:
    s = input().rstrip()
    if s == "#":
        break
    _, key = s.split()
    bst.insert(bst.root, int(key))
bst.preOrder()
```

```python
class ListNode:
    def __init__(self, val, next=None):
        self.val = val
        self.next = next


class LinkedList:
    def __init__(self, val) -> None:
        self.head = ListNode(val)
        self.tail = self.head
        self.keys = {val}

    def addFirst(self, val):
        if val in self.keys:
            return
        node = ListNode(val)
        node.next = self.head
        self.head = node
        self.keys.add(val)

    def addLast(self, val):
        if val in self.keys:
            return
        node = ListNode(val)
        self.tail.next = node
        self.tail = node
        self.keys.add(val)

    def reverse(self):
        prev = None
        curr = self.head
        while curr:
            curr.next, prev, curr = prev, curr, curr.next
        self.head = prev

    def printList(self):
        node = self.head
        while node:
            print(node.val, end=" ")
            node = node.next

    def addAfter(self, val, key):
        if key not in self.keys or val in self.keys:
            return
        self.keys.add(val)
        node = ListNode(val)
        temp = self.head
        while temp:
            if temp.val == key:
                node.next = temp.next
                temp.next = node
                return
            temp = temp.next

    def addBefore(self, val, key):
        if key not in self.keys or val in self.keys:
            return
        self.keys.add(val)
        node = ListNode(val)
        if self.head.val == key:
            node.next = self.head
            self.head = node
            return
        temp = self.head
        while temp.next:
            if temp.next.val == key:
                node.next = temp.next
                temp.next = node
                return
            temp = temp.next

    def remove(self, val):
        if val not in self.keys:
            return
        self.keys.remove(val)
        if self.head.val == val:
            self.head = self.head.next
            return
        temp = self.head
        while temp.next:
            if temp.next.val == val:
                temp.next = temp.next.next
                if not temp.next:
                    self.tail = temp
                return
            temp = temp.next


import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline
n = int(input())
a = list(map(int, input().split()))
ll = LinkedList(a[0])
ll.keys.add(a[0])
for i in a[1:]:
    ll.addLast(i)
    ll.keys.add(i)
    # ll.tail = node
while True:
    s = input().rstrip()
    if s == "#":
        break
    if s == "reverse":
        ll.reverse()
    else:
        s = s.split()
        if s[0] == "addlast":
            ll.addLast(int(s[1]))
        elif s[0] == "addfirst":
            ll.addFirst(int(s[1]))
        elif s[0] == "addafter":
            ll.addAfter(int(s[1]), int(s[2]))
        elif s[0] == "addbefore":
            ll.addBefore(int(s[1]), int(s[2]))
        elif s[0] == "remove":
            ll.remove(int(s[1]))
ll.printList()
```
