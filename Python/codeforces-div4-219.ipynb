```python
import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


class Solution:
    def volume(self, h, a, x):
        s = 0
        for i in a:
            s += max(h - i, 0)
        return s <= x


t = int(input())
s = Solution()
for _ in range(t):
    n, x = map(int, input().split())
    a = list(map(int, input().split()))
    r = max((sum(a) + x) // n, max(a))
    l = 1
    while l + 1 < r:
        m = (l + r) // 2
        if s.volume(m, a, x):
            l = m
        else:
            r = m
    print(r if s.volume(r, a, x) else l)
```

```python
import sys
from collections import deque

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


t = int(input())
for _ in range(t):
    n, k = map(int, input().split())
    a = list(map(int, input().split()))
    h = list(map(int, input().split()))
    if all(x > k for x in a):
        print(0)
        continue
    a.append(0)
    h.append(1)
    dq = deque()
    fruits = 0
    max_len = 1
    for i in range(n):
        if h[i] % h[i + 1] == 0:
            dq.append(i)
            fruits += a[i]
            if fruits > k:
                max_len = max(max_len, len(dq) - 1)
                while fruits > k:
                    fruits -= a[dq.popleft()]
        else:
            fruits += a[i]
            max_len = max(max_len, len(dq) if fruits > k else len(dq) + 1)
            fruits = 0
            dq.clear()
    max_len = max(max_len, len(dq))
    print(max_len)
```

```python
import sys

sys.stdin = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.inp", "r")
sys.stdout = open(r"C:\Users\nguye\Documents\Code\Programming\C++\test.out", "w")
input = sys.stdin.readline


t = int(input())
for _ in range(t):
    s = input().rstrip()
    if s.count("A") == 0 or s.count("B") == 0:
        print(0)
        continue
    index_B = [-1]
    for i in range(len(s)):
        if s[i] == "B":
            index_B.append(i)
    index_B.append(len(s))
    ans = 0
    mins = float("inf")
    for i in range(len(index_B) - 1):
        ans += index_B[i + 1] - index_B[i] - 1
        mins = min(mins, index_B[i + 1] - index_B[i] - 1)
    print(ans - mins)
```
