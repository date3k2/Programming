```python
import sys
from collections import defaultdict
from bisect import bisect_left, bisect_right

# sys.stdin = open('input.txt', 'r')
# sys.stdout = open('output.txt', 'w')
input = sys.stdin.readline


def to_time(time: str):
    h, m, s = map(int, time.split(":"))
    return 3600 * h + 60 * m + s


# --- Analyze sales order of an e-commerce company ---
if __name__ == "__main__":
    total_number_orders = 0
    total_revenue = 0
    revenue_of_shop = defaultdict(int)
    total_consume_of_customer_shop = defaultdict(int)
    time_revenues = []
    while True:
        s = input().rstrip()
        if s == "#":
            break
        customer_id, product_id, price, shop_id, time = map(str, s.split())
        price = int(price)
        total_revenue += price
        total_number_orders += 1
        revenue_of_shop[shop_id] += price
        total_consume_of_customer_shop[customer_id + shop_id] += price
        time_revenues.append([to_time(time), price])
    time_revenues.sort()
    for i in range(1, total_number_orders):
        time_revenues[i][1] = time_revenues[i - 1][1] + time_revenues[i][1]
    times = [time_revenues[i][0] for i in range(total_number_orders)]
    revenues = [time_revenues[i][1] for i in range(total_number_orders)]
    count_line = 0
    while True:
        query = input().rstrip()
        if query == "#":
            break
        if query == "?total_number_orders":
            print(total_number_orders)
        elif query == "?total_revenue":
            print(total_revenue)
        else:
            query = query.split()
            if len(query) == 2:
                _, shop_id = query
                print(revenue_of_shop[shop_id])
            elif query[0] == "?total_consume_of_customer_shop":
                _, customer_id, shop_id = query
                print(total_consume_of_customer_shop[customer_id + shop_id])
            else:
                _, start, end = query
                start = to_time(start)
                end = to_time(end)
                if start > end or start > times[-1] or end < times[0]:
                    print(0)
                    continue
                l = bisect_left(times, start)
                r = bisect_right(times, end)
                print(revenues[r - 1] - revenues[l - 1] if l > 0 else revenues[r - 1])
```
